```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

bar_width = 0.5

a = np.random.normal(size=10000)
bins = np.arange(-3, 3, step=bar_width)
hist, edges = np.histogram(a, bins)
bins = edges[:-1]

fig, ax = plt.subplots(figsize=(12, 6))
ax.bar(bins, hist/len(a), ec='w', lw=.5, width=bar_width, align='edge')
plt.show()
```
