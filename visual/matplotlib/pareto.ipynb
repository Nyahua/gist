# Twin X, or Y axis
just make another axes to share the x or y axis (`ax.twinx/y`)

https://en.wikipedia.org/wiki/Pareto_index

```python
x = com = [chr(i)+" Ltd." for i in np.arange(10) + ord("A")]
y = np.sort(np.random.pareto(1.16, 10))[::-1]
y_per = np.cumsum((y / y.sum()))

fig, ax = plt.subplots(figsize=(11, 7))
ax.bar(np.arange(len(x)), y, width=0.95)
ax.set_xticks(np.arange(len(x)))
ax.set_xticklabels(x)
ax.xaxis.set_ticks_position('none')

ax2 = ax.twinx()
ax2.plot(y_per, marker="o", color='r')
ax2.set_yticks(np.linspace(0, 1, 5))
labels = ["{:.0%}".format(i) for i in np.linspace(0, 1, 5)]
ax2.set_yticklabels(labels)
ax2.set_ylim(0, 1.1)

plt.show()
```
