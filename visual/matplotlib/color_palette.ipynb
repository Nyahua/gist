```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
nums = 6
bottoms = np.arange(6)
values = np.sort(100 + np.random.normal(scale=30, size=6))


colors = np.repeat('#50514F', nums)
colors[-2] = "#247BA0"

```

```python
fig, ax = plt.subplots(figsize=(9, 6))

ax.barh(bottoms, values, color=colors)
ax.set_axis_off()
plt.show()
```

```python
values_1 = values + np.sort(np.abs(np.random.normal(10, 30, nums)))
fig, ax = plt.subplots(figsize=(9, 6))

ax.plot(values, color='#50514F', lw=2)
ax.plot(values_1, color="#F25F5C", lw=2)
ax.set_axis_off()
plt.show()
```
