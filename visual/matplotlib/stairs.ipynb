```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# CONSTANT
DARK_BLUE = "#08519c"
DARK_GRAY = "#252525"
LIGHT_GRAY = "#969696"
```

```python
nums = 5
box_label = ["STEP "+str(i+1) for i in range(nums)]

xmin, xmax = -1, nums+0.5
ymin, ymax = -1, nums+0.5
box_height, box_width = 0.8, 1

fig, ax = plt.subplots(figsize=(12, 9))

for idx in range(5):
    ax.scatter(idx, idx, s=800, color=DARK_BLUE)
    ax.text(idx, idx, str(idx+1), 
            ha='center', va="center", color='w', fontsize=14)
    ax.bar(idx+0.5, -box_height, box_width, idx, color=DARK_GRAY, zorder=0)
    ax.text(idx+0.5, idx-box_height/2, box_label[idx],
            ha='center', va="center", color='w', fontsize=16)
    
ax.arrow(-1, -1, nums+1, nums+1, head_width=0.1,
         head_length=0.2, fc=LIGHT_GRAY, ec=LIGHT_GRAY, zorder=0)

ax.arrow(xmin, ymin, nums+1.5, 0, head_width=0.1,
         head_length=0.2, fc=DARK_GRAY, ec=DARK_GRAY, zorder=0)
ax.arrow(xmin, ymin, 0, nums+1.5, head_width=0.1,
         head_length=0.2, fc=DARK_GRAY, ec=DARK_GRAY, zorder=0)

ax.hlines(range(nums+1), xmin, xmax, color=LIGHT_GRAY, lw=0.75, zorder=0)
ax.vlines(range(nums+1), ymin, ymax, color=LIGHT_GRAY, lw=0.75, zorder=0)

ax.set_xlim(xmin-0.1, xmax+0.5)
ax.set_ylim(ymin-0.1, ymax+0.5)

ax.set_axis_off()
plt.show()
```
