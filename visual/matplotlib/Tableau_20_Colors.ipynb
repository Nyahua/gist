# Tableau 20 Colors

```python
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
import pandas as pd

# These are the "Tableau 20" colors as RGB & HEX.  
tableau20 = ['#%02x%02x%02x' %  rgb for rgb in [(31, 119, 180), (174, 199, 232), (255, 127, 14), (255, 187, 120),  
             (44, 160, 44), (152, 223, 138), (214, 39, 40), (255, 152, 150), (148, 103, 189), (197, 176, 213), 
             (140, 86, 75), (196, 156, 148),  (227, 119, 194), (247, 182, 210), (127, 127, 127), (199, 199, 199),  
             (188, 189, 34), (219, 219, 141), (23, 190, 207), (158, 218, 229)] ]

fig, ax = plt.subplots(1,1, figsize=(9, 6))

v, h = np.arange(20) // 2, np.arange(20)%2
bar_band = ax.barh(v, np.ones(20), height=1, left=h)

for i in range(20):
    bar_band[i].set_color(tableau20[i])
    text_color = 'k' if i%2 else 'w'
    ax.text(h[i]+.05, v[i], i, ha='left', va='center', color=text_color)
    ax.text(h[i]+.35, v[i], tableau20[i], ha='right', va='center', family='monospace', color=text_color)
    
ax.invert_yaxis()
ax.set_axis_off()
    
plt.show()
```

```python
colors = [(31, 119, 180), (174, 199, 232), (255, 127, 14), (255, 187, 120),  
            (44, 160, 44), (152, 223, 138), (214, 39, 40), (255, 152, 150), (148, 103, 189), (197, 176, 213), 
            (140, 86, 75), (196, 156, 148),  (227, 119, 194), (247, 182, 210), (127, 127, 127), (199, 199, 199),  
            (188, 189, 34), (219, 219, 141), (23, 190, 207), (158, 218, 229)]
lines = ["  {}  {}  {}  {}\n".format(r, g, b, '#%02x%02x%02x'%(r,g,b)) for (r, g, b) in colors]

filename = 'tableau_20.gpl'
with open(filename, 'w') as f:
    
    f.write("GIMP Palette\n")
    f.write("Name: Tableau 20 Colors\n")
    f.write("# generated by PaletteGen.py\n")

    for line in lines:
        f.write(line)
        
f.close()
```
