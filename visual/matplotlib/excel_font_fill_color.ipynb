```python
import pandas as pd
import numpy as np
import openpyxl as opx
```

```python
excel_filename = r'siemens_ase.xlsx'
df = pd.read_excel(excel_filename)
hexs = df["hex"]
rgbs = df.apply(lambda x: (x['R'], x['G'], x['B']), axis=1)
del df
```

```python
def rgb_font_color(rgb):
    l = (np.array(rgb) * np.array([0.299, 0.587, 0.114])).sum()
    font_color = "#000000" if l > 186 else "#ffffff"
    return font_color
```

```python
wb = opx.load_workbook(excel_filename)
ws = wb[wb.sheetnames[0]]
for idx in range(85):
    row = idx + 2
    fill = opx.styles.PatternFill(fill_type="solid", start_color=hexs[idx])
    font = opx.styles.Font(color=rgb_font_color(rgbs[idx])[-6:])
    cell = ws.cell(row=row, column=5)
    cell.fill = fill 
    cell.font = font
    
    
wb.save(excel_filename)
```
