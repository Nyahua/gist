```python
import json
import folium
from pathlib import Path
import branca
```

```python
geo_path = Path("geojson")
with open(geo_path / 'heilongjiang.json', encoding='utf-8') as f:
    json_hlj = json.loads(f.read())
```

```python
# Create a white image of 4 pixels, and embed it in a url.
white_tile = branca.utilities.image_to_url([[1, 1], [1, 1]])
cp = json_hlj['cp'][::-1]

# Create a map using this url for each tile.
m = folium.Map(
    location=cp, 
    zoom_start=6, 
    tiles=white_tile, 
    attr='white tile'
)
# chropleth by city
folium.Choropleth(
    geo_data=json_hlj,
    fill_color='red',
    fill_opacity=0.3,
    line_weight=2,
    line_color='white'
).add_to(m)

# city markers
cities = json_hlj['features']
for city in cities:
    name, cp = city['properties']['name'], city['properties']['cp'][::-1]

    folium.map.Marker(
        location = cp,
        icon=folium.features.DivIcon(
            icon_size=(150,36),
            icon_anchor=(30,10),
            html=f'<div style="font-size: 10pt">{name}</div>',
            )
        ).add_to(m)

m
```
