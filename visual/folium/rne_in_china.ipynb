```python
import folium
import branca
from pathlib import Path
import numpy as np
import json


#SIEMENS COLORS CONSTANT
TONE_DARK = "#3c464b"
STONE_LIGHT = "#879baa"
STONE_LIGHT_35 = "#becdd7"
SAND_DARK = "#73645a"
SAND_LIGHT = "#aaaa96"
SAND_LIGHT_35 = "#d7d7cd"
TEAL_DARK = "#00646e"
TEAL_LIGHT = "#41aaaa"
BLUE_DARK = "#005f87"
BLUE_LIGHT = "#50bed7"
YELLOW_DARK = "#eb780a"
YELLOW_LIGHT = "#ffb900"
RED_DARK = "#641946"
RED_LIGHT = "#af235f"
GREEN_DARK = "#647d28"
GREEN_LIGHT = "#aab414"

# Create a white image of 4 pixels, and embed it in a url.
white_tile = branca.utilities.image_to_url([[1, 1], [1, 1]])
# Create a map using this url for each tile.
m = folium.Map(
    tiles=white_tile, 
    attr='white tile',
)

def choropleth(geojson, fill_color='green', fill_opacity=1, line_weight=2, line_color='white'):
    # chropleth by city
    folium.Choropleth(
        geo_data=geojson,
        fill_color=fill_color,
        fill_opacity=fill_opacity,
        line_weight=line_weight,
        line_color=line_color
    ).add_to(m)
    
    return m

geo_path = Path("geojson")
with open(geo_path / 'china.json', encoding='utf-8') as f:
    json_china = json.loads(f.read())
with open(geo_path / 'lnn_ime.json', encoding='utf-8') as f:
    json_lnn_ime = json.loads(f.read())
with open(geo_path / 'lns.json', encoding='utf-8') as f:
    json_lns = json.loads(f.read())
with open(geo_path / 'heilongjiang.json', encoding='utf-8') as f:
    json_heilongjiang = json.loads(f.read())
with open(geo_path / 'jilin.json', encoding='utf-8') as f:
    json_jilin = json.loads(f.read())
```

```python
choropleth(json_china, fill_color=STONE_LIGHT_35, line_weight=1, line_color=STONE_LIGHT_35)
choropleth(json_lnn_ime, fill_color=BLUE_LIGHT, line_weight=2, line_color=BLUE_LIGHT)
choropleth(json_lns, fill_color=RED_LIGHT, line_weight=2, line_color=RED_LIGHT)
choropleth(json_jilin, fill_color=TEAL_LIGHT, line_weight=2, line_color=TEAL_LIGHT)
choropleth(json_heilongjiang, fill_color=GREEN_LIGHT, line_weight=2, line_color=GREEN_LIGHT)
m.fit_bounds(m.get_bounds())
m
```

```python
m.save('ren_in_china.html')
```
