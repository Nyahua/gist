```python
import json
import folium
from pathlib import Path
import branca
```

```python
geo_path = Path("geojson")
with open(geo_path / 'jilin.json', encoding='utf-8') as f:
    json_jilin = json.loads(f.read())
```

```python
# Create a white image of 4 pixels, and embed it in a url.
white_tile = branca.utilities.image_to_url([[1, 1], [1, 1]])
cp = json_jilin['cp'][::-1]

# Create a map using this url for each tile.
m = folium.Map(
    location=cp, 
    zoom_start=7, 
    tiles=white_tile, 
    attr='white tile'
)
# chropleth by city
folium.Choropleth(
    geo_data=json_jilin,
    fill_color='red',
    fill_opacity=0.3,
    line_weight=2,
    line_color='white'
).add_to(m)

# city markers
cities = json_jilin['features']
for city in cities:
    name, cp = city['properties']['name'], city['properties']['cp'][::-1]
    html=f"""
    <p>{name}</p>
    <a href="#">link</a>
    """
    iframe = folium.IFrame(html, figsize=(2, 2))
    popup = folium.Popup(iframe)
    
    folium.CircleMarker(
        radius=5,
        location=cp,
        popup=popup,
        color='white',
        fill=True,
    ).add_to(m)

m
```
